```python
# A big example of everything
reload(onl)
# Draw a string of elements
d = schem.Drawing(unit=2)
len_ct = 2*0.75

ee = d.add(onl.BUS)
ee.add_label('A', loc='top')
ee.add_label('105', loc='bot')

ee = d.add(onl.CT, d='right', l=len_ct)
ee.add_label('B', loc='top')
ee.add_label('110', loc='bot')
ee = d.add(onl.CT, d='right', l=len_ct)
ee.add_label('C', loc='top')
ee.add_label('115', loc='bot')

bb = d.add(onl.SWITCH, l=len_ct)
bb.add_label('F', loc='top')
bb.add_label('120', loc='bot')
bb = d.add(onl.BREAKER, l=len_ct)
bb.add_label('D', loc='top')
bb.add_label('120', loc='bot')
bb = d.add(onl.SWITCH, l=len_ct)
bb.add_label('F', loc='top')
bb.add_label('120', loc='bot')

ee = d.add(onl.CT, d='right', l=len_ct)
ee.add_label('C', loc='top')
ee.add_label('125', loc='bot')
ee = d.add(onl.CT, d='right', l=len_ct)
ee.add_label('B', loc='top')
ee.add_label('130', loc='bot')

bb = d.add(onl.SUBCOND)
bb.add_label('E', loc='top')
bb.add_label('135', loc='bot')

# The location of the DOT is important because we want to draw 
# out from this point later.
dot = d.add(e.DOT)
dot_loc = dot.start

d.add(e.LINE, d='up', l=2*d.unit)
bb = d.add(onl.SUBCOND, d='left')
bb.add_label('E', loc='top')
bb.add_label('205', loc='bot')
# This keeps going left for a ways

d.add(e.LINE, d='right', xy=dot_loc, l=0.5)
bb = d.add(onl.WAVETRAP)
bb.add_label('G', loc='top')
bb.add_label('305', loc='bot')
# 
d.draw()

```
